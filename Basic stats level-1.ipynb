```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv(r"D:\Downloads\Basic stats - 1\Basic stats - 1\sales_data_with_discounts.csv")

# Identify numerical and categorical columns
numerical_cols = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
categorical_cols = ['Date', 'Day', 'SKU', 'City', 'BU', 'Brand', 'Model']

# ---------------------------
# Descriptive Statistics
# ---------------------------
mean_vals = df[numerical_cols].mean()
median_vals = df[numerical_cols].median()
std_vals = df[numerical_cols].std()
mode_vals = df[numerical_cols].mode().iloc[0] 

# Combine into one DataFrame
stats_summary = pd.DataFrame({
    'Mean': mean_vals,
    'Median': median_vals,
    'Mode': mode_vals,
    'Standard Deviation': std_vals
})
print("Descriptive Statistics:\n", stats_summary)

# ---------------------------
# Histograms and Boxplots
# ---------------------------
sns.set(style="whitegrid")
plt.figure(figsize=(16, 20))

for i, col in enumerate(numerical_cols):
    plt.subplot(6, 2, 2*i+1)
    sns.histplot(df[col], kde=True, bins=30, color="skyblue")
    plt.title(f'Histogram of {col}')

    plt.subplot(6, 2, 2*i+2)
    sns.boxplot(x=df[col], color="salmon")
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()

# ---------------------------
# Bar Charts for Categorical Columns
# ---------------------------
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(y=df[col], order=df[col].value_counts().index, palette="Set2")
    plt.title(f'Count Plot of {col}')
    plt.tight_layout()
    plt.show()

# ---------------------------
# Standardization (Z-score)
# ---------------------------
scaler = StandardScaler()
standardized_data = pd.DataFrame(scaler.fit_transform(df[numerical_cols]), columns=numerical_cols)

print("\nStandardized Numerical Columns (first 5 rows):")
print(standardized_data.head())

# Optional: Compare original and standardized distributions using histograms
for col in numerical_cols:
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, color="skyblue")
    plt.title(f'Original Distribution: {col}')

    plt.subplot(1, 2, 2)
    sns.histplot(standardized_data[col], kde=True, color="green")
    plt.title(f'Standardized Distribution: {col}')

    plt.tight_layout()
    plt.show()

# ---------------------------
# One-Hot Encoding for Categorical Variables
# ---------------------------
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
print("\nDataset with Dummy Variables (first 5 rows):")
print(df_encoded.head())
```
